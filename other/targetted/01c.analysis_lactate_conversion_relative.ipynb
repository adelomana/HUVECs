```python
import pandas, numpy
import scipy, scipy.stats

import matplotlib, matplotlib.pyplot
matplotlib.rcParams.update({'font.size':20, 'font.family':'FreeSans', 'xtick.labelsize':20, 'ytick.labelsize':20})
```

# 1. read data

```python
df = pandas.read_csv(data_file, header = 1)
df.head()
```

# 2. format data

## 2.1. select appropriate columns

```python
selected_columns = [0, 1, 2, 3, 4] + [7, 8]
df = df.iloc[:, selected_columns]
df
```

## 2.1. select appropriate rows

```python
df.drop([0], inplace=True)
df.drop(list(range(141, 171+1)), inplace=True)
```

```python
wdf = df[df['Cell numbers'].str.isnumeric() == True]
wdf['Cell numbers'] = wdf['Cell numbers'].astype(float)
```

```python
wdf['Glucose'][wdf['Glucose'] == '-'] = float('nan')
wdf['Lactate'][wdf['Lactate'] == '-'] = float('nan')
```

```python
wdf['Glucose'] = wdf['Glucose'].astype(float)
wdf['Lactate'] = wdf['Lactate'].astype(float)

wdf['Treatment'] = wdf['Treatment'].astype(float)
wdf['Timepoint (h)'] = wdf['Timepoint (h)'].astype(float)
```

```python
pandas.set_option("display.max_rows", 200, "display.max_columns", 20)
print(wdf)
```

```python
unique_experiments = list(set(wdf['Experiment'].tolist()))
unique_experiments.sort()
unique_experiments
```

# 3. analysis

```python
the_colors = ['blue', 'green', 'pink', 'orange', 'red']
quant = []
for time in [4, 24]:
    for treatment in [0, 0.5]:
        box = []
        for experiment in unique_experiments:
            
            if time == 4:
                time_reference = 0
            elif time == 24:
                time_reference = 4
            else:
                raise ValueError('unrecognized time value')
                
            ###
            ### cell numbers for both metabolites
            ###
                
            cn_current = wdf[(wdf['Treatment'] == treatment) & (wdf['Timepoint (h)'] == time) & (wdf['Experiment'] == experiment)]['Cell numbers'].astype(float).values
            cn_reference = wdf[(wdf['Treatment'] == treatment) & (wdf['Timepoint (h)'] == time_reference) & (wdf['Experiment'] == experiment)]['Cell numbers'].astype(float).values
            
            cnc = 1e6/cn_current
            cnr = 1e6/cn_reference
            
            ###
            ### glucose
            ###
            
            glucose_fvs = wdf[(wdf['Treatment'] == treatment) & (wdf['Timepoint (h)'] == time) & (wdf['Experiment'] == experiment)]['Glucose'].astype(float).values
            glucose_fvs_reference = wdf[(wdf['Treatment'] == treatment) & (wdf['Timepoint (h)'] == time_reference) & (wdf['Experiment'] == experiment)]['Glucose'].astype(float).values
            
            glu_current = glucose_fvs*cnc
            glu_reference = glucose_fvs_reference*cnr
            
            if len(glu_reference) == 0:
                pass
            elif len(glu_reference) == 1:
                glu_consumption = glu_current - glu_reference[0]
            else:
                glu_consumption = glu_current - glu_reference
                        
            ###
            ### lactate
            ###
            
            lactate_fvs = wdf[(wdf['Treatment'] == treatment) & (wdf['Timepoint (h)'] == time) & (wdf['Experiment'] == experiment)]['Lactate'].astype(float).values
            lactate_fvs_reference = wdf[(wdf['Treatment'] == treatment) & (wdf['Timepoint (h)'] == time_reference) & (wdf['Experiment'] == experiment)]['Lactate'].astype(float).values
            
            lac_current = lactate_fvs*cnc
            lac_reference = lactate_fvs_reference*cnr
            
            if len(lac_reference) == 0:
                pass
            elif len(lac_reference) == 1:
                lac_secretion = lac_current - lac_reference[0]
            else:
                lac_secretion = lac_current - lac_reference
            
            ###
            ### ratio
            ###
            
            ratio = abs(lac_secretion/glu_consumption)
                        
            if time == 4:
                if experiment == '2' or experiment == '6':
                    
                    quant.append(ratio)
            
                    print(the_colors[unique_experiments.index(experiment)])
                    print('treatment', treatment, 'time', time)
                    print('glu', glu_current, glu_reference)
                    print('lac', lac_current, lac_reference)
                    print('glu consumption', glu_consumption)
                    print('lac secretion', lac_secretion)
                    print('ratio', ratio)
                    print()
                    
            if time == 24:
                if experiment == '2' or experiment == '5':
                    
                    quant.append(ratio)
            
                    print(the_colors[unique_experiments.index(experiment)])
                    print('treatment', treatment, 'time', time)
                    print('glu', glu_current, glu_reference)
                    print('lac', lac_current, lac_reference)
                    print('glu consumption', glu_consumption)
                    print('lac secretion', lac_secretion)
                    print('ratio', ratio)
                    print()
```

```python
# blue and orange references
a = quant[0]/numpy.mean(quant[0])
b = quant[1]/numpy.mean(quant[1])
print(b)

# blue and orange treatments
c = quant[2]/numpy.mean(quant[0])
d = quant[3]/numpy.mean(quant[1])

### 24 h
# blue and pink references
e = quant[4]/numpy.mean(quant[4])
f = quant[5]/numpy.mean(quant[5])

# blue and pink treatments
g = quant[6]/numpy.mean(quant[4])
h = quant[7]/numpy.mean(quant[5])

relatives = [a, b, c, d, e, f, g, h]
```

 # 4. plot

```python
the_figure = matplotlib.pyplot.figure()
ax = the_figure.add_subplot(111)

###
### boxplots
####

bx = matplotlib.pyplot.boxplot(relatives, showcaps=False, medianprops=dict(linewidth=2, color='black'), boxprops=dict(linewidth=2, color='gray'), whiskerprops=dict(linewidth=2, color='gray'))

###
### scatter plots
###

the_colors = ['blue', 'orange', 'blue', 'orange', 'blue', 'pink', 'blue', 'pink']
x=[]; y=[]; exp_colors=[]
for i in range(len(relatives)):
    for element in relatives[i]:
  
        shift = 0.25
        r=-shift/2 + shift*numpy.random.random()
        x.append(i+1+r)

        y.append(element)

        exp_colors.append(the_colors[i])

matplotlib.pyplot.scatter(x, y, s=80, c=exp_colors, alpha=0.8, linewidths=0)

###
### statistics
###

stat, pvalue = scipy.stats.mannwhitneyu(c, a)
print(stat, pvalue)
stat, pvalue = scipy.stats.mannwhitneyu(d, b)
print(stat, pvalue)

stat, pvalue = scipy.stats.mannwhitneyu(g, e)
print(stat, pvalue)
stat, pvalue = scipy.stats.mannwhitneyu(h, f)
print(stat, pvalue)

matplotlib.pyplot.text(7, 4.05, '*', horizontalalignment='center', verticalalignment='center', fontsize=40)
matplotlib.pyplot.text(8, 4.05, '*', horizontalalignment='center', verticalalignment='center', fontsize=40)

###
### close figure
###

matplotlib.pyplot.grid(alpha=1/3)

matplotlib.pyplot.axvline(4.5, alpha=0.5, lw=2, color='black', ls=':')
matplotlib.pyplot.text(0.25, 0.05, 'T=4', transform=ax.transAxes, horizontalalignment='center', verticalalignment='center')
matplotlib.pyplot.text(0.75, 0.05, 'T=24', transform=ax.transAxes, horizontalalignment='center', verticalalignment='center')

matplotlib.pyplot.xticks([1, 2, 3, 4, 5, 6, 7, 8], ['0', '0', '0.5', '0.5', '0', '0', '0.5', '0.5'])
matplotlib.pyplot.yticks([1, 2, 3, 4])

matplotlib.pyplot.xlabel('Treatment (uM)')
matplotlib.pyplot.ylabel('Lac/Glu ratio relative change')

matplotlib.pyplot.tight_layout()
matplotlib.pyplot.show()
```
